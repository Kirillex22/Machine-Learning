# Подготовка данных

Рассмотрим уникальные значения в столбце Sunshine, характеризующем количество солнечных часов.

```python
data['Sunshine'].unique()
```

Поскольку среди уникальных значений присутствует 0, NaN не означает отсутствие солнца как такового.

Рассмотрим датасет без пропусков:

```python
temp = data.dropna()
```

```python
temp
```

Отсюда видно, что даже в подряд идущих днях значения в Sunshine не коррелируют. Обратим внимание на гистограмму распределения данной величины

```python
temp.hist('Sunshine', bins = len(data['Sunshine'].unique()))
```

Отсюда видно, что большинство дней были пасмурными. Поскольку заполнение средним значением может сильно исказить данные, заполним пропуски модой, то есть нулем.

```python
data['Sunshine'] = data['Sunshine'].fillna(0)
```

```python
data.isna().sum().sort_values()
```

Аналогично рассмотрим уникальные значения величины Evaporation (кол-ва осадков) и их распределение 

```python
data['Evaporation'].unique()
```

```python
temp.hist('Evaporation', bins = len(data['Evaporation'].unique()))
```

Найдем моду и заполним ей пропуски:

```python
evap_mode = data['Evaporation'].mode().iat[0]
print(evap_mode)
```

```python
data['Evaporation'] = data['Evaporation'].fillna(evap_mode)
```
